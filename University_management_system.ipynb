```python
import random

def generate_otp():
    return random.randint(100000, 999999)

class Person:
    def __init__(self, id, name, email):
        self.name = name
        self.id = id
        self.email = email

class Student(Person):
    def __init__(self, id, name, email, branch):
        super().__init__(id, name, email)
        self.branch = branch

class Teacher(Person):
    def __init__(self, id, name, email, subject):
        super().__init__(id, name, email)
        self.subject = subject

class College:
    def __init__(self, id, name):
        self.id = id
        self.name = name
        self.students = []
        self.teachers = []

    def enroll_student(self, student):
        self.students.append(student)

    def hire_teacher(self, teacher):
        self.teachers.append(teacher)

    def find_students_by_branch(self, branch):
        return [s for s in self.students if s.branch == branch]

    def find_teachers_by_subject(self, subject):
        return [t for t in self.teachers if t.subject == subject]

institutions = []

while True:
    print("Choose an Option")
    print("1. Register College")
    print("2. Register Student")
    print("3. Register Teacher")
    print("4. Show All Students")
    print("5. Show All Teachers")
    print("6. List Students by Branch")
    print("7. List Teachers by Subject")
    print("8. Exit")

    choice = int(input("Enter Your Option: "))

    if choice == 1:
        cname = input("Enter College Name: ")
        cid = input("Enter College ID: ")
        if any(col.id == cid for col in institutions):
            print("************************")
            print("College Already Registered!")
            print("************************")
        else:
            institutions.append(College(cid, cname))
            print("************************")
            print("College Registered Successfully")
            print("************************")

    elif choice == 2:
        cid = input("Enter College ID: ")
        clg = next((col for col in institutions if col.id == cid), None)

        if clg:
            name = input("Enter Student Name: ")
            roll = input("Enter Roll Number: ")
            email = input("Enter Email: ")
            branch = input("Enter Branch: ")

            otp = generate_otp()
            print(f"OTP for verification: {otp}")
            if int(input("Enter the OTP: ")) == otp:
                clg.enroll_student(Student(roll, name, email, branch))
                print("*************************")
                print("Student Registered Successfully!")
                print("*************************")
            else:
                print("*************************")
                print("Invalid OTP! Registration Failed.")
                print("*************************")
        else:
            print("************************")
            print("College Not Found!")
            print("************************")

    elif choice == 3:
        cid = input("Enter College ID: ")
        clg = next((col for col in institutions if col.id == cid), None)

        if clg:
            name = input("Enter Teacher Name: ")
            roll = input("Enter Employee ID: ")
            email = input("Enter Email: ")
            subject = input("Enter Subject: ")

            otp = generate_otp()
            print(f"OTP for verification: {otp}")
            if int(input("Enter the OTP: ")) == otp:
                clg.hire_teacher(Teacher(roll, name, email, subject))
                print("*************************")
                print("Teacher Registered Successfully!")
                print("*************************")
            else:
                print("*************************")
                print("Invalid OTP! Registration Failed.")
                print("*************************")
        else:
            print("************************")
            print("College Not Found!")
            print("************************")

    elif choice == 4:
        cid = input("Enter College ID: ")
        clg = next((col for col in institutions if col.id == cid), None)

        if clg and clg.students:
            print("**********************************")
            print(f"Students of {clg.name}: ")
            for s in clg.students:
                print(f"ID: {s.id}, Name: {s.name}, Email: {s.email}, Branch: {s.branch}")
            print("**********************************")
        else:
            print("************************")
            print("No Students Found!")
            print("************************")

    elif choice == 5:
        cid = input("Enter College ID: ")
        clg = next((col for col in institutions if col.id == cid), None)

        if clg and clg.teachers:
            print("**********************************")
            print(f"Teachers of {clg.name}: ")
            for t in clg.teachers:
                print(f"ID: {t.id}, Name: {t.name}, Email: {t.email}, Subject: {t.subject}")
            print("**********************************")
        else:
            print("************************")
            print("No Teachers Found!")
            print("************************")

    elif choice == 6:
        cid = input("Enter College ID: ")
        branch = input("Enter Branch: ")
        clg = next((col for col in institutions if col.id == cid), None)

        if clg:
            students = clg.find_students_by_branch(branch)
            if students:
                print(f"Students in {branch} branch at {clg.name}: ")
                for s in students:
                    print(f"ID: {s.id}, Name: {s.name}, Email: {s.email}")
            else:
                print("************************")
                print("No Students Found in this Branch.")
                print("************************")
        else:
            print("************************")
            print("College Not Found!")
            print("************************")

    elif choice == 7:
        cid = input("Enter College ID: ")
        subject = input("Enter Subject: ")
        clg = next((col for col in institutions if col.id == cid), None)

        if clg:
            teachers = clg.find_teachers_by_subject(subject)
            if teachers:
                print(f"Teachers for {subject} at {clg.name}: ")
                for t in teachers:
                    print(f"ID: {t.id}, Name: {t.name}, Email: {t.email}")
            else:
                print("************************")
                print("No Teachers Found for this Subject.")
                print("************************")
        else:
            print("************************")
            print("College Not Found!")
            print("************************")

    elif choice == 8:
        print("************************")
        print("Thank You for Using the System!")
        print("************************")
        break

    else:
        print("************************")
        print("Invalid Option! Please Try Again.")
        print("************************")
```
